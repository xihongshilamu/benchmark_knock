```python
import scanpy as sc
import pandas as pd
import numpy as np
# some standard packages to assist this tutorial
import matplotlib.pyplot as plt
import pandas as pd
import scanpy as sc
import anndata as ad
from scipy.sparse import csr_matrix
```

```python
data_c = pd.read_csv('/data/share/data/data_c_del_rowcol.csv')
data_c.set_index('Gene Name', drop=True, append=False, inplace=True)
data_c = np.log(data_c + 1)
data_t = pd.read_csv('/data/share/data/data_t_del_rowcol.csv')
data_t.set_index('Gene Name', drop=True, append=False, inplace=True)
data_t = np.log(data_t + 1)
print(data_c.shape, data_t.shape)
```

```python
columns_name = list(data_t.columns)
print(len(columns_name))
data_c.columns = columns_name
# 将要敲除的基因值置为 0
data_c_copy = data_c.copy()
for i in data_c.columns:
    ko_name = i.split('_')[0]
    ko_name = ko_name.split('/')[0]
    data_c.at[ko_name, i] = 0
```

```python
data_c = data_c.drop(columns=['-----_707'])
data_t = data_t.drop(columns=['-----_707'])
columns_name = list(data_t.columns)
```

```python
data_test_c1 = pd.read_csv('./data_c_choose_3000.csv')

data_test_c1.set_index('Gene Name', drop=True, append=False, inplace=True)
name_test = list(data_test_c1.index)

data_c=data_c.loc[name_test]

data_t=data_t.loc[name_test]
name_drop = data_t.index
data_c = data_c.values.T
data_t = data_t.values.T
```

```python
data_c.shape, data_t.shape
```

```python
# knock data c
print(len(columns_name))
obs = pd.DataFrame()
obs['batch'] = np.random.choice(['0', '1'], data_c.shape[0]) # 样本数，每个样本观测信息
obs['knock_gene'] = columns_name
obs['knock_or_not'] = 'yes'
var_names = name_test
n_vars = len(var_names)         # 输出大小为genes数目
var = pd.DataFrame(index=var_names)
print(len(columns_name)) # 敲除的基因list
new_data_c = ad.AnnData(data_c,obs=obs,var=var)

```

```python
# unknock data t
obs = pd.DataFrame()
obs['batch'] = np.random.choice(['0', '1'], data_t.shape[0]) # 样本数，每个样本观测信息
obs['knock_gene'] = columns_name
obs['knock_or_not'] = 'no'
var_names = name_test
n_vars = len(var_names)         # 输出大小为genes数目
var = pd.DataFrame(index=var_names)
print(len(columns_name)) # 敲除的基因list
new_data_t = ad.AnnData(data_t,obs=obs,var=var)

```
